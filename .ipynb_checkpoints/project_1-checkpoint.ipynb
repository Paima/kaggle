**Функция предобработки данных**

```python
def data_clean(df):
    
    df['ap_hi'] = df['ap_hi'].where((df['ap_hi']>0),df['ap_hi']*(-1))
    df['ap_lo'] = df['ap_lo'].where((df['ap_lo']>0),df['ap_lo']*(-1))
    #df = df.drop(df.loc[df['ap_hi']<60].index)
    #df = df.drop(df.loc[df['ap_hi']>300].index)
    #df = df.drop(df.loc[df['ap_lo']<30].index)
    #df = df.drop(df.loc[df['ap_lo']>150].index)
    #df = df.drop(df.loc[df['height']<135].index)
    #df = df.drop(df.loc[df['height']>200].index)
    #df = df.drop(df.loc[df['weight']<30].index)
    #df = df.drop(df.loc[df['weight']>150].index)
    #df = df.drop(df.loc[df['age']<14000].index)
    #df = df.drop(df.loc[df['age']>24000].index)
    df['ap_hi'] = df['ap_hi'].where((df['ap_hi']>60),60)
    df['ap_lo'] = df['ap_lo'].where((df['ap_lo']>30),30)
    df['ap_hi'] = df['ap_hi'].where((df['ap_hi']<300),300)
    df['ap_lo'] = df['ap_lo'].where((df['ap_lo']<150),150)
    df['height'] = df['height'].where((df['height']>135),135)
    df['height'] = df['height'].where((df['height']<200),200)
    df['weight'] = df['weight'].where((df['weight']>30),30)
    df['weight'] = df['weight'].where((df['weight']<155),155)
    df['age'] = df['age'].where((df['age']>14000),14000)
    df['age'] = df['age'].where((df['age']<24000),24000)
    return df

```

## Разработка модели предсказания риска заболеваний сердца

Посчитаем Классы в целевом признаке

```python
class_frequency = data_train['cardio'].value_counts(normalize=True)
print(class_frequency)
```

Дисбаланса практически нет. Константная модель выдаст результат 49.5%

Разобьем тренировочные данные на три выборки: тренировочную, валидационную и тестовую

```python
#???
#data_ohe = pd.get_dummies(data_train, drop_first=True)
features = data_train.drop(columns=['cardio'])
target = data_train['cardio']
```

```python
features
```

```python
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.25, random_state=12345)
```

```python
features_train.describe()
```

```python
#Численные
features_train_sc = features_train
features_valid_sc = features_valid
#scaler = StandardScaler()
#scaler.fit(features_train_sc[numeric])
#features_train_sc[numeric] = scaler.transform(features_train_sc[numeric])
#features_valid_sc[numeric] = scaler.transform(features_valid_sc[numeric]) 
#features_valid_sc
```

```python
best_lr_model = None
best_lr_result = 0
best_lr_iter = 0
auc_roc_lr = 0
best_f1_lr = 0

for iters in range(100,110):
    model_lr = LogisticRegression(random_state=12345, solver='lbfgs', max_iter=iters)
    model_lr.fit(features_train_sc,target_train)
    predicted_valid = model_lr.predict(features_valid_sc)
    result = accuracy_score(target_valid, predicted_valid)
    cur_f1 = f1_score(target_valid, predicted_valid)
    probabilities_valid = model_lr.predict_proba(features_valid_sc)
    probabilities_one_valid = probabilities_valid[:, 1]
    auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
    if best_lr_result < result:
        auc_roc_lr = auc_roc
        best_f1_lr = cur_f1
        best_lr_result = result
        best_lr_model = model_lr
        best_lr_iter = iters
print('acc:',best_lr_result,'\niters:', best_lr_iter,'\nAUC-ROC:', auc_roc_lr, '\nbest_f1:',best_f1_lr)
```

```python
best_tree_model = None
best_tree_result = 0
best_tree_depth = 0
auc_roc_tree = 0
best_f1_tree = 0


for depth in range(5,10):
    model_tree = DecisionTreeClassifier(random_state=12345,max_depth=depth)
    model_tree.fit(features_train_sc,target_train)
    predicted_valid = model_tree.predict(features_valid_sc)
    result = accuracy_score(target_valid, predicted_valid)
    cur_f1 = f1_score(target_valid, predicted_valid)
    probabilities_valid = model_tree.predict_proba(features_valid_sc)
    probabilities_one_valid = probabilities_valid[:, 1]
    auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
    if best_tree_result < result:
        auc_roc_tree = auc_roc
        best_f1_tree = cur_f1
        best_tree_result = result
        best_tree_model = model_tree
        best_tree_depth = depth

print('acc:',best_tree_result,'\ndepth:', best_tree_depth,'\nAUC-ROC:', auc_roc_tree, '\nbest_f1:',best_f1_tree)
```

```python
best_f1_rf = 0
best_est = 0
best_rf_depth = 0
auc_roc_rf = 0
best_model_rf = None
best_rf_result = 0

#hyperopt???
for est in [100,200,300,400,500,600,700]:
    for depth in [1,3,5,7,9,11,13,15]:
            model = RandomForestClassifier(n_estimators=est, random_state=12345, max_depth=depth,n_jobs=-1)
            model.fit(features_train_sc, target_train)
            predicted_valid = model.predict(features_valid_sc)
            cur_f1 = f1_score(target_valid, predicted_valid)
            probabilities_valid = model.predict_proba(features_valid_sc)
            probabilities_one_valid = probabilities_valid[:, 1]
            auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
            result = accuracy_score(target_valid, predicted_valid)
            if best_rf_result < result:
                auc_roc_rf = auc_roc
                best_f1_rf = cur_f1
                best_est = est
                best_rf_depth = depth
                best_model_rf = model
                best_rf_result = result
print('acc:',best_rf_result,'\ndepth:', best_rf_depth,'\nAUC-ROC:', auc_roc_rf, '\nbest_f1:',best_f1_rf)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
best_f1_gb = 0
best_est = 0
best_depth_gb = 0
auc_roc_gb = 0
best_model_gb = None
best_gb_result = 0

for est in [100,200,300,400]:
    for depth in [1,3,5,7]:
            model = GradientBoostingClassifier(n_estimators=est, random_state=12345, max_depth=depth)
            model.fit(features_train_sc, target_train)
            predicted_valid = model.predict(features_valid_sc)
            cur_f1 = f1_score(target_valid, predicted_valid)
            probabilities_valid = model.predict_proba(features_valid_sc)
            probabilities_one_valid = probabilities_valid[:, 1]
            auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
            result = accuracy_score(target_valid, predicted_valid)
            if best_gb_result < result:
                auc_roc_gb = auc_roc
                best_f1_gb = cur_f1
                best_est = est
                best_depth_gb = depth
                best_model_gb = model
                best_gb_result = result
print('acc:',best_gb_result,'\ndepth:', best_depth_gb,'\nAUC-ROC:', auc_roc_gb, '\nbest_f1:',best_f1_gb)
```
